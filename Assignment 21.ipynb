.Question 1-write a python program to scrape data for "Data Analyst" job position in "Bangalore" location.You have to scrape the job-title,job-location,company_name,experiance_required.you have to scrape first 10 job data.

```python
#Q1.webpage
url = 'https://www.naukri.com/'  
driver.get(url)

#Enter "Data Analyst" in "Skill,Designations,Comapnies" field
search_job = driver.find_element_by_id('qsb-keyword-sugg')
search_job.send_keys("Data Analyst")


#enter"Bangalore"in "enter the location"field
search_loc = driver.find_element_by_id('qsb-location-sugg')
search_loc.send_keys("Bangalore")

#Click the search button
search_btn = driver.find_element_by_xpath("//button[@class='btn']")
search_btn.click()

#scrape the job-title,job_location,campany_name,experiance_requried empty list.
job_titles=[]
campany_name=[]
experiance_list=[]
location_list=[]

#scrape the job-title
titles_tags= driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']")
titles_tags
job_titles=[]
for i in titles_tags:
    job_titles.append(i.text)
job_titles

#top 10 job list
titles_tags[:10]


#scrape the campany_name.
companies_tags=driver.find_elements_by_xpath("//a[@class='subTitle ellipsis fleft']")
companies_tags
company_names=[]

for i in companies_tags:
    company_names.append(i.text)
company_names


#scrape the experiance_requried.
experiance_tags=driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi experience']")
experiance_tags
experiance_list=[]
for i in experiance_tags:
    experiance_list.append(i.text)
experiance_list

#scrape the job_location.
locations_tags=driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi location']//span")
locations_tags
locations_list=[]
for i in locations_tags:
    locations_list.append(i.text)
locations_list

print(len(job_titles),len(company_names),len(experiance_list),len(locations_list))

#DataFrame
jobs=pd.DataFrame({})
jobs['title']=job_titles
jobs['campany']=company_names
jobs['experiance_requried']=experiance_list
jobs['location']=locations_list

```
